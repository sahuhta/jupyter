```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('StudentsPerformance.csv')
mat = data['math score']
luk = data['reading score']
plt.scatter(x=mat, y=luk)
plt.title('Matikan ja lukutaidon arvosanat')
plt.xlabel('matikka')
plt.ylabel('lukutaito')
```

```python
mat = data['writing score']
luk = data['reading score']
plt.scatter(x=mat, y=luk)
plt.title('Kirjoitus- ja lukutaidon arvosanat')
plt.xlabel('kirjoitustaito')
plt.ylabel('lukutaito')
```
